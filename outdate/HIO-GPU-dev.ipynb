# Build reciprocal space data

```python
@jit
def reciprocal_mask(x, y, z, r):
    
    local_mask = np.zeros((128,)*3 )
    
    R1 = r**2
    R2 = (r+2)**2
    count = 0
    
    for l in range(128):
        for m in range(128):
            for n in range(128):
                condition = (((x-l)**2 + (y - m)**2 + (z - n)**2 >= R1) &
                            ((x-l)**2 + (y - m)**2 + (z - n)**2 <= R2) )
                
                if condition :
                    count += 1
                    local_mask[l,m,n] = 1
                    
                condition = (((128 -x-l)**2 + (128 -y - m)**2 + (128 -z - n)**2 >= R1) &
                            ((128 -x-l)**2 + (128 -y - m)**2 + (128 -z - n)**2 <= R2) )
                
                if condition :
                    count += 1
                    local_mask[l,m,n] = 1
    
    return count , local_mask        
```

```python
pixel_number, momentum_mask = reciprocal_mask(64. , 64., 0., 63. )
print pixel_number
```

```python
measurement = np.sqrt(np.multiply(momentum_mask , intensity))


f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize=(15,5))
ax1.imshow(np.sum(measurement,axis=0),vmax=70)
ax2.imshow(np.sum(measurement,axis=1),vmax=70)
ax3.imshow(np.sum(measurement,axis=2),vmax=70)
```

# Phase retrieval

```python
#diffract = np.fft.fftshift(np.sqrt(intensity))

diffract = np.fft.fftshift(np.copy(measurement))
momentum_mask = np.fft.fftshift(momentum_mask)

#mask = np.fft.fftshift(mask)
```

```python
import time

#Initial guess using random phase info
guess = diffract * np.exp(1j * np.random.rand(128,128,128) * 2 * pi)
update = np.zeros_like(guess)

#number of iterations
r = 1001

#step size parameter
beta = 0.8

#Finishes the first round because it's different from the other round
prev = None

# Begin the loop
time1 = time.time()

for s in range(0,r):
    #apply fourier domain constraints
    update[momentum_mask==1] = diffract[momentum_mask==1] * np.exp(1j * np.angle(guess[momentum_mask==1]))
    #update = diffract * np.exp(1j * np.angle(guess))
    
    inv = fft.ifftn(update)
    inv = np.real(inv)
    if prev is None:
        prev = inv
        
    #apply real-space constraints
    temp = inv
    
    ## This part need to be vectorize
    inv[(inv<0)&(mask==1)] = prev[(inv<0)&(mask==1)] - beta* inv[(inv<0)&(mask==1)]
    inv[mask==0] = prev[mask ==0 ] - beta*inv[mask ==0]
    
    #inv *= mask
    #inv[(inv<0)] = 0
    
    prev = temp
    
    guess = fft.fftn(inv)
        
    #save an image of the progress
    if s % 50 == 0:
        
        time2 = time.time()
        np.save("./3d_save_single/progress" + str(s) + ".npy", prev)
        print s/50, time2-time1
        
```
